# EDA - East West Airlines

East West Airlines busca mejorar la efectividad de una campana comercial realizada en alianza con una empresa de telecomunicaciones. Para ello, se utilizara informacion historica de sus clientes, como millas acumuladas, actividad reciente, uso de tarjetas, estatus elite y pertenencia al club, con el fin de estimar que clientes tienen mayor probabilidad de comprar el servicio Telcom ofrecido en la campana. Esto permitira segmentar mejor la base de clientes, priorizar esfuerzos comerciales y optimizar recursos de marketing y ventas.

## Enfoque del proyecto

Este proyecto analiza la base de clientes de East West Airlines para entender que variables se relacionan con la compra del servicio Telcom (`phone_sale`).

La base permite describir perfiles, comparar tasas de conversion observadas y preparar informacion para un modelo predictivo. No se asumen etapas comerciales, eventos de campana ni reglas automaticas que no esten registradas en los datos.

Cuando una idea dependa de informacion no disponible, se presenta como recomendacion futura y no como condicion aplicada al analisis.

```python
# Librerias
import matplotlib.pyplot as plt
import seaborn as sns
import re
import unicodedata
import pandas as pd
import numpy as np
```

```python
# Carga del dataset
df = pd.read_csv('../input/EastWestAirlinesNN.csv')

# Copia de respaldo para comparar contra la data original
df_raw = df.copy()

# Vista inicial del dataset cargado
print('Filas y columnas originales:', df_raw.shape)
display(df_raw.head())
```

## 1. Validacion y estandarizacion de nombres de columnas

```python
# Auditoria de nombres originales
columnas_originales = pd.DataFrame({
    'posicion': range(1, len(df.columns) + 1),
    'nombre_original': df.columns
})

# Mostrar nombres antes de cambiarlos
display(columnas_originales)
```

```python
# Funcion para estandarizar nombres de columnas
def estandarizar_nombre_columna(nombre):
    nombre = str(nombre).strip().lower()
    nombre = unicodedata.normalize('NFKD', nombre)
    nombre = ''.join(c for c in nombre if not unicodedata.combining(c))
    nombre = re.sub(r'[^a-z0-9]+', '_', nombre)
    nombre = re.sub(r'_+', '_', nombre).strip('_')
    return nombre

# Crear nombres limpios
columnas_estandarizadas = [estandarizar_nombre_columna(col) for col in df.columns]
```

```python
# Comparacion entre nombres originales y nombres limpios
auditoria_columnas = pd.DataFrame({
    'nombre_original': df.columns,
    'nombre_estandarizado': columnas_estandarizadas,
    'cambio': df.columns != pd.Index(columnas_estandarizadas)
})

# Validar que no se generen columnas duplicadas
if pd.Series(columnas_estandarizadas).duplicated().any():
    raise ValueError('La estandarizacion genero nombres duplicados.')

# Aplicar nombres estandarizados
df.columns = columnas_estandarizadas
display(auditoria_columnas)
print(df.columns.tolist())
```

Los nombres originales tenian caracteres como `#` y `?`. Se estandarizan para evitar errores al llamar columnas durante el analisis.

## 2. Unidad de analisis y granularidad

```python
# Definir columnas clave del caso
id_col = 'id'
target_col = 'phone_sale'

# Resumen de granularidad
granularidad = pd.DataFrame({
    'indicador': [
        'filas',
        'columnas',
        'ids no nulos',
        'ids unicos',
        'ids duplicados',
        'filas sin id',
        'filas sin variable principal'
    ],
    'valor': [
        df.shape[0],
        df.shape[1],
        df[id_col].notna().sum(),
        df[id_col].nunique(dropna=True),
        df[id_col].duplicated().sum(),
        df[id_col].isna().sum(),
        df[target_col].isna().sum()
    ]
})

# Mostrar hallazgo de granularidad
display(granularidad)
```

**Hallazgo:** la base esta a nivel cliente: `id` tiene un valor unico por registro valido. Hay 2 filas sin `id` y sin `phone_sale`, por lo que se revisan en nulos antes de continuar.

## 3. Revision y correccion de tipos de datos

Primero se revisa el tipo detectado por Python para cada columna. El objetivo es corregir formatos antes de calcular estadisticas o graficos.

```python
# Revision inicial de tipos y valores unicos
tipos_iniciales = pd.DataFrame({
    'columna': df.columns,
    'tipo_inicial': df.dtypes.astype(str),
    'nulos': df.isna().sum().values,
    'valores_unicos': df.nunique(dropna=True).values
})

# Mostrar diagnostico de tipos
display(tipos_iniciales)
```

```python
# Definir columnas esperadas para aplicar conversiones de tipo
binary_cols = [
    'topflight', 'cc1_miles', 'cc2_miles', 'cc3_miles',
    'email', 'club_member', 'any_cc_miles_12mo', 'phone_sale'
]

count_cols = ['bonus_trans', 'flight_trans_12', 'online_12']
miles_cols = ['balance', 'qual_miles', 'bonus_miles', 'flight_miles_12mo']
numeric_cols = binary_cols + count_cols + miles_cols
```

```python
# Convertir objetos a string si aparecen en el archivo
object_cols = df.select_dtypes(include='object').columns.tolist()
for col in object_cols:
    df[col] = df[col].astype('string')

# Convertir identificador a string porque no es una variable cuantitativa
df[id_col] = pd.to_numeric(df[id_col], errors='coerce').astype('Int64').astype('string')

# Convertir columnas numericas a valores numericos
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convertir binarias y conteos a enteros que aceptan nulos
for col in binary_cols + count_cols:
    df[col] = df[col].round().astype('Int64')
```

```python
# Revision final de tipos corregidos
tipos_corregidos = pd.DataFrame({
    'columna': df.columns,
    'tipo_corregido': df.dtypes.astype(str),
    'nulos': df.isna().sum().values,
    'valores_unicos': df.nunique(dropna=True).values
})

# Mostrar resultado de conversion
display(tipos_corregidos)
```

No se encontraron columnas de fecha ni columnas categoricas de texto. El identificador del cliente se convirtio a `string` porque no representa una cantidad. Las variables de millas se conservaron como numericas y las variables binarias/conteos se dejaron como enteros que aceptan nulos (`Int64`).

## 4. Validacion de variables binarias

Segun el diccionario de variables, no hay columnas categoricas de texto. `id` es un identificador, no una categoria de analisis. En esta seccion solo se valida que las variables binarias del dataset mantengan valores esperados: 0 y 1.

```python
# Validar que las binarias solo tengan valores 0 o 1
validacion_binarias = []

for col in binary_cols:
    valores = sorted(df[col].dropna().unique().tolist())
    valores_no_validos = [v for v in valores if v not in [0, 1]]
    validacion_binarias.append({
        'columna': col,
        'valores_observados': valores,
        'estado': 'OK' if not valores_no_validos else 'Revisar'
    })

# Mostrar validacion de variables binarias
display(pd.DataFrame(validacion_binarias))
```

Las variables binarias del dataset tienen valores dentro de lo esperado

## 5. Valores nulos e imputacion

```python
# Resumen de nulos por columna
nulos_antes = pd.DataFrame({
    'columna': df.columns,
    'nulos': df.isna().sum().values,
    'porcentaje_nulos': (df.isna().mean().values * 100).round(2)
}).sort_values('nulos', ascending=False)

# Mostrar columnas con nulos
display(nulos_antes)
```

**Hallazgo:** los nulos aparecen en el identificador, en la variable de compra del servicio Telcom y en algunas variables del cliente. Como la compra del servicio Telcom es la variable objetivo, se revisan esas filas antes de decidir cualquier imputacion.

```python
# Detectar filas con muy poca informacion
filas_muy_vacias = df.isna().mean(axis=1) >= 0.80
filas_sin_clave = df[id_col].isna() | df[target_col].isna()
filas_a_eliminar = filas_muy_vacias | filas_sin_clave

# Mostrar filas candidatas a eliminacion
print('Filas a eliminar:', int(filas_a_eliminar.sum()))
display(df.loc[filas_a_eliminar])
```

**Decision:** se eliminan las 2 filas sin identificador y sin resultado de compra del servicio Telcom. No se imputa la variable objetivo porque seria inventar si el cliente compro o no compro.

```python
# Eliminar filas sin informacion clave
df = df.loc[~filas_a_eliminar].copy()

# Preparar tabla de imputacion si quedaran nulos
imputacion = []

# Imputar solo los nulos restantes
for col in df.columns:
    nulos = int(df[col].isna().sum())
    if nulos == 0:
        continue
    if pd.api.types.is_numeric_dtype(df[col]):
        valor = df[col].median()
        metodo = 'mediana'
    else:
        valor = df[col].mode(dropna=True).iloc[0]
        metodo = 'moda'
    df[col] = df[col].fillna(valor)
    imputacion.append({'columna': col, 'nulos_imputados': nulos, 'metodo': metodo, 'valor': valor})

# Mostrar resultado del tratamiento de nulos
display(pd.DataFrame(imputacion) if imputacion else pd.DataFrame({'resultado': ['No quedaron nulos por imputar.']}))
print('Shape despues de tratar nulos:', df.shape)
```

## 6. Duplicados

La revision de duplicados se realiza despues del tratamiento de nulos para no contar filas vacias.

```python
# Calcular duplicados completos y duplicados por cliente
duplicados_completos = int(df.duplicated().sum())
duplicados_id = int(df.duplicated(subset=[id_col]).sum())

# Mostrar resultado
print('Duplicados completos:', duplicados_completos)
print('Duplicados por id:', duplicados_id)
```

```python
# Eliminar duplicados completos si existieran
if duplicados_completos > 0:
    df = df.drop_duplicates().copy()
    print('Se eliminaron duplicados completos.')
else:
    print('No se eliminaron filas por duplicidad completa.')

# Revisar clientes repetidos por id si existieran
if duplicados_id > 0:
    display(df[df.duplicated(subset=[id_col], keep=False)].sort_values(id_col))
else:
    print('No hay clientes repetidos por id.')
```

La base final queda sin duplicados completos y sin clientes repetidos por `id`.

```python
# Variables seleccionadas para los analisis descriptivos posteriores
numeric_analysis_cols = ['balance', 'bonus_miles', 'flight_miles_12mo', 'bonus_trans', 'flight_trans_12', 'online_12']
binary_analysis_cols = ['topflight', 'email', 'club_member', 'any_cc_miles_12mo', 'cc1_miles', 'cc2_miles', 'cc3_miles']

# Etiquetas de negocio basadas en el diccionario de variables
variable_labels = {
    'id': 'Identificador del cliente',
    'topflight': 'Estatus elite Topflight',
    'balance': 'Millas disponibles para premios',
    'qual_miles': 'Millas calificables para Topflight',
    'cc1_miles': 'Uso de tarjeta frequent flyer',
    'cc2_miles': 'Uso de tarjeta Rewards',
    'cc3_miles': 'Uso de tarjeta Small Business',
    'bonus_miles': 'Millas por bonos no-vuelo',
    'bonus_trans': 'Transacciones bono no-vuelo',
    'flight_miles_12mo': 'Millas voladas ultimos 12 meses',
    'flight_trans_12': 'Vuelos ultimos 12 meses',
    'online_12': 'Compras online ultimos 12 meses',
    'email': 'Email registrado',
    'club_member': 'Miembro del club de la aerolinea',
    'any_cc_miles_12mo': 'Acumulo millas con alguna tarjeta',
    'phone_sale': 'Compra del servicio Telcom'
}

# Mantener alias para reutilizar el resto del notebook
important_numeric_cols = numeric_analysis_cols
important_binary_cols = binary_analysis_cols
```

# 7. Outliers

Antes de interpretar distribuciones, correlaciones o cruces con la compra del servicio Telcom, se revisan valores extremos con IQR. En este negocio, un valor extremo puede ser un cliente premium o muy activo, por lo que se evalua antes de eliminar.

```python
# Variables para analisis de outliers
outlier_cols = ['balance', 'qual_miles', 'bonus_miles', 'bonus_trans', 'flight_miles_12mo', 'flight_trans_12', 'online_12']
outliers = []

# Calcular limites IQR por variable
for col in outlier_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    limite_inferior = q1 - 1.5 * iqr
    limite_superior = q3 + 1.5 * iqr
    mascara = (df[col] < limite_inferior) | (df[col] > limite_superior)
    outliers.append({
        'variable': variable_labels[col],
        'outliers': int(mascara.sum()),
        'porcentaje_outliers': round(float(mascara.mean() * 100), 2),
        'limite_superior': limite_superior,
        'maximo': df[col].max()
    })

# Mostrar resumen de outliers
outliers = pd.DataFrame(outliers).sort_values('porcentaje_outliers', ascending=False)
display(outliers)
```

```python
# Boxplots con transformacion log1p para visualizar outliers
fig, axes = plt.subplots(2, 4, figsize=(17, 7))
axes = axes.flatten()

# Graficar una variable por panel
for ax, col in zip(axes, outlier_cols):
    sns.boxplot(x=np.log1p(df[col]), ax=ax)
    ax.set_title(variable_labels[col])
    ax.set_xlabel('Escala log1p')

# Ocultar paneles no usados
for ax in axes[len(outlier_cols):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

**Hallazgo y decision:** los outliers mas relevantes aparecen en millas voladas en los ultimos 12 meses (16.0%), transacciones de vuelo recientes (12.8%), saldo de millas (7.5%) y millas bono (7.4%). Se conservan porque pueden representar clientes de alto valor. Esta decision no modifica la base; solo advierte que medias, histogramas y correlaciones deben interpretarse considerando esa concentracion.

# 8. Analisis Univariado

Se revisa cada variable por separado para entender tamano, concentracion y frecuencia antes de relacionarla con la compra del servicio Telcom. Como los outliers se conservaron, las medianas y porcentajes se interpretan junto con las medias para no sobredimensionar a los clientes extremos.

## 8.1 Estadistica descriptiva de variables numericas

Se revisan saldo de millas, millas bono, vuelos recientes, transacciones y actividad online para identificar concentracion y dispersion en el comportamiento de los clientes.

```python
# Estadistica descriptiva de variables numericas
resumen_univariado_num = df[numeric_analysis_cols].describe().T

# Agregar medidas adicionales
resumen_univariado_num['mediana'] = df[numeric_analysis_cols].median()
resumen_univariado_num['moda'] = df[numeric_analysis_cols].mode().iloc[0]
resumen_univariado_num['asimetria'] = df[numeric_analysis_cols].skew()

# Mostrar resumen con nombres de negocio
resumen_univariado_num.index = resumen_univariado_num.index.map(variable_labels)
display(resumen_univariado_num)
```

**Hallazgo:** las variables de millas y vuelos estan muy concentradas: muchos clientes tienen poca o ninguna actividad reciente, mientras pocos acumulan saldos y vuelos muy altos. Estos clientes extremos no deben eliminarse de entrada porque pueden representar viajeros frecuentes o clientes de alto valor.

```python
# Histogramas univariados de variables numericas
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

# Graficar una variable por panel
for ax, col in zip(axes, numeric_analysis_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=ax)
    ax.set_title(variable_labels[col])
    ax.set_xlabel(variable_labels[col])
    ax.set_ylabel('Clientes')

plt.tight_layout()
plt.show()
```

## 8.2 Frecuencias de variables binarias

Se revisan las frecuencias absolutas y porcentajes de variables 0/1, como estatus elite, email registrado, pertenencia al club y uso de tarjetas asociadas a millas.

```python
# Calcular frecuencias de variables binarias
frecuencias = []

for col in binary_analysis_cols:
    conteo = df[col].value_counts(dropna=False).sort_index()
    porcentaje = (conteo / len(df) * 100).round(2)
    for valor in conteo.index:
        frecuencias.append({
            'variable': variable_labels[col],
            'valor': valor,
            'clientes': conteo.loc[valor],
            'porcentaje': porcentaje.loc[valor]
        })

# Mostrar tabla de frecuencias
frecuencias_binarias = pd.DataFrame(frecuencias)
display(frecuencias_binarias)
```

```python
# Graficos de barras univariados para variables binarias
fig, axes = plt.subplots(2, 4, figsize=(16, 7))
axes = axes.flatten()

# Graficar frecuencia por variable
for ax, col in zip(axes, binary_analysis_cols):
    sns.countplot(data=df, x=col, ax=ax)
    ax.set_title(variable_labels[col])
    ax.set_xlabel('0 = no, 1 = si')
    ax.set_ylabel('Clientes')

# Ocultar paneles vacios
for ax in axes[len(binary_analysis_cols):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## 8.3 Distribucion univariada de la variable principal

La variable principal indica si el cliente compro el servicio Telcom como resultado de la campana. Antes de cruzarla con otras variables, se revisa sola para identificar el balance entre compradores y no compradores.

```python
# Distribucion de la variable principal
target_dist = (
    df[target_col]
    .value_counts()
    .rename_axis('compra_telcom')
    .reset_index(name='clientes')
)

# Calcular porcentaje
target_dist['resultado'] = target_dist['compra_telcom'].map({0: 'No compro', 1: 'Compro'})
target_dist['porcentaje'] = (target_dist['clientes'] / len(df) * 100).round(2)
target_dist = target_dist[['resultado', 'clientes', 'porcentaje']]

# Mostrar distribucion
display(target_dist)
```

```python
# Grafico de barras de la variable principal
fig, ax = plt.subplots(figsize=(6, 4))
sns.countplot(data=df, x=target_col, ax=ax)

# Etiquetas del grafico
ax.set_title('Distribucion de compra del servicio Telcom')
ax.set_xlabel('Compra del servicio Telcom (0 = no, 1 = si)')
ax.set_ylabel('Clientes')
plt.tight_layout()
plt.show()
```

**Hallazgo:** De 4,985 clientes, 1,042 compraron el servicio Telcom, lo que representa una tasa de compra de 20.9%. Como los compradores son una minoria frente a los no compradores, el modelo debera evaluarse con metricas adecuadas para clases desbalanceadas y no solo con accuracy.

# 10. Analisis Bivariado

Luego de revisar relaciones generales entre variables, se cruza la compra del servicio Telcom con caracteristicas del cliente. El objetivo es encontrar senales comerciales previas al modelo: que perfiles compran mas y que variables conviene evaluar como predictoras.

## 10.1 Compra del servicio Telcom vs variables binarias

Se compara la tasa de compra entre clientes con y sin cada indicador: estatus elite, email, club y uso de tarjetas asociadas a millas.

```python
# Funcion para calcular tasa de compra por grupo
def tasa_compra_por(variable):
    tabla = (
        df.groupby(variable, dropna=False)[target_col]
        .agg(clientes='count', tasa_compra_telcom='mean')
        .reset_index()
    )
    tabla['tasa_compra_telcom'] = (tabla['tasa_compra_telcom'] * 100).round(2)
    tabla['variable'] = variable_labels[variable]
    tabla['valor'] = tabla[variable]
    return tabla[['variable', 'valor', 'clientes', 'tasa_compra_telcom']]

# Calcular cruces bivariados con variables binarias
tablas_cruce = []
for variable in binary_analysis_cols:
    tablas_cruce.append(tasa_compra_por(variable))

# Mostrar tabla consolidada
bivariado_binarias = pd.concat(tablas_cruce, ignore_index=True)
display(bivariado_binarias)
```

```python
# Graficos bivariados de tasa de compra por grupo
fig, axes = plt.subplots(2, 2, figsize=(13, 8))
variables_grafico = ['topflight', 'email', 'club_member', 'any_cc_miles_12mo']
axes = axes.flatten()

# Graficar tasa de compra para variables binarias clave
for ax, variable in zip(axes, variables_grafico):
    tabla = tasa_compra_por(variable)
    sns.barplot(data=tabla, x='valor', y='tasa_compra_telcom', ax=ax)
    ax.set_title(variable_labels[variable])
    ax.set_xlabel('0 = no, 1 = si')
    ax.set_ylabel('% compra Telcom')

plt.tight_layout()
plt.show()
```

**Hallazgo:** la pertenencia al club es la senal mas fuerte: los 457 miembros registran 100% de compra del servicio Telcom. Tambien destacan el estatus elite (32.1% vs 18.6% en clientes sin ese estatus), el uso de alguna tarjeta para acumular millas (25.3% vs 15.8%) y la tarjeta frequent flyer (25.4% vs 16.0%). Para el modelo, estas variables capturan fidelizacion y vinculo comercial con la aerolinea.

## 10.1.1 Cruces observados de clientes

En lugar de crear segmentos accionables con reglas asumidas, se revisan combinaciones observadas de variables existentes. Esto permite comparar tasas de compra Telcom sin convertir esas combinaciones en etiquetas comerciales artificiales.

```python
# Cruces observados con variables disponibles, sin crear segmentos inventados
cruces_observados = (
    df.groupby(['email', 'club_member', 'topflight'])[target_col]
    .agg(clientes='count', tasa_conversion_telcom='mean')
    .reset_index()
)
cruces_observados['porcentaje_base'] = (cruces_observados['clientes'] / len(df) * 100).round(2)
cruces_observados['tasa_conversion_telcom'] = (cruces_observados['tasa_conversion_telcom'] * 100).round(2)
cruces_observados = cruces_observados.sort_values('tasa_conversion_telcom', ascending=False)

display(cruces_observados)
```

**Lectura:** estos cruces solo muestran diferencias observadas en la base. No definen clientes potenciales, prioridades ni acciones automaticas; para eso harian falta reglas de negocio o datos de campana validados.

## 10.2 Compra del servicio Telcom vs variables numericas

Se compara el nivel de millas, transacciones y actividad reciente entre clientes que compraron y clientes que no compraron el servicio Telcom.
